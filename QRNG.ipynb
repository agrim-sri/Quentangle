```python
import json
import requests
import subprocess
import os

QRN_URL = "https://qrng.anu.edu.au/API/jsonI.php"
QRN_KEY = os.environ.get("QRN_API_KEY") or "YOUR_API_KEY_IF_SECURE"

DTYPE = "hex16"
LENGTH = 640
BLOCKSIZE = 1

headers = {"x-apiKey": QRN_KEY}
params = {"length": LENGTH, "type": DTYPE, "size": BLOCKSIZE}


# fetch random data
response = requests.get(QRN_URL, headers=headers, params=params) 

if response.status_code != 200:
    print(f"unexpected status-code: {response.status_code}")
    exit()

js = response.json()
if not js["success"]:
    print(js["message"])
    exit()

random_seed = "".join(map(str, js["data"]))

test_result = subprocess.check_output(['wg'], input=random_seed.encode())

print(random_seed)
print("Randomness test result:")
print(test_result.decode())

num_clients = int(input("Enter the number of clients: "))
client_details = []

for i in range(num_clients):
    name = input(f"Enter name for client {i + 1}: ")
    email = input(f"Enter email for client {i + 1}: ")

    private_key = subprocess.check_output(["wg", "genkey"], input=random_seed.encode()).decode().strip()
    public_key = subprocess.check_output(["wg", "pubkey"], input=private_key.encode()).decode().strip()

    client_details.append((name, email, private_key, public_key))

print("\nClient details:")
for i, (name, email, private_key, public_key) in enumerate(client_details):
    print(f"\nClient {i + 1}:")
    print(f"Name: {name}")
    print(f"Email: {email}")
    print(f"Private key: {private_key}")
    print(f"Public key: {public_key}")
```
